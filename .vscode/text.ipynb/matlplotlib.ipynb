```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn 
```

```python
arr = np.random.randint(20000,40000,100)
arr
```

```python
#lineplot
arr.sort()
plt.plot(arr)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [10,11,12,45,67]
plt.plot(x,y,marker = "o" ,ms = 10,
         mfc = "red",mec = "green",
         color = "black",
         lw =3,ls =":",
         alpha = 0.9,label = "lineplot")
plt.title("lineplt")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()

plt.show()
```

```python
#scatterplot
plt.scatter(x,y,marker="d")
plt.show()
```

```python
x = ["mp","up","rj","gh"]
y = [10,20,30,40]
colours = ["red","black","blue","green"]
hatch_patterns = ["//","\\\\","x","--"]
plt.bar(x,y,color = colours,hatch = hatch_patterns,
        label = x,width =0.3,edgecolor = "black",alpha = 0.7,
        linewidth = 3)
plt.legend()
#plt.grid()
plt.show()
```

```python
x = ["mp","up","rj","gh"]
y = [10,20,30,40]
colours = ["red","black","blue","green"]
hatch_patterns = ["//","\\\\","x","--"]
#plt.bar(x,y,color = colours,hatch = hatch_patterns,
        #label= x,width =0.3,edgecolor = "black",alpha = 0.7,
        #linewidth = 3)
plt.barh(x,y)        
plt.legend()
#plt.grid()
plt.show()
```

```python
##histogram -- distribution n of data in interval (bins)
arr = np.random.randint(2000,40000,100)
plt.hist(arr,bins = 20)
plt.show()
```

```python
##piechart --o perctentage 
x = ["mp","up","rj","gj"]
y = [10,20,30,40]
explodes = (0.2,0,0,0)
plt.pie(y,labels = x,autopct="%1.2f%%",shadow="True",explode= explodes)
plt.legend()
#plt.savefig("piechaart.png") #kisi bhi format me ho jayega 
plt.show
```

```python
##subplot

x = np.linspace(0,10,100)
plt.Figure(figsize=(10,6))
plt.subplot(2,2,1) ## 2 rows and 2 coloumns,1st graph
plt.plot(x,np.sin(x))
plt.title("sin")
plt.subplot(2,2,2) ## 2 rows and 2 coloumns,2nd graph
plt.plot(x,np.cos(x))
plt.title("cos")
plt.subplot(2,2,3) ## 2 rows and 2 coloumns,3rd graph
plt.plot(x,np.tan(x))
plt.title("tan")
plt.subplot(2,2,4) ## 2 rows and 2 coloumns,4th graph
plt.plot(x,np.log(x))
plt.title("log")
plt.tight_layout()
plt.show()
```
