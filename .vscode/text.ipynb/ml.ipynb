```python
# Define the features (x) and target (y)
x = final_df.drop(columns=['target_column'])  # Drop the target column from the features
y = final_df['target_column']  # Define the target column
```

```python
## traing and testing
from sklearn.model_selection  import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x , y , test_size = 0.2,random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
#y_pred = lr.predict(x_test)
```

```python
y_pred = lr.predict(x_test)
y_pred
```

```python
from sklearn import metrics
print('mse:',metrics.mean_absolute_error(y_test,y_pred))
print('MSE:',metrics.mean_squared_error(y_test,y_pred))
print('rmse:',np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
plt.scatter(y_test,y_pred)
plt.plot(y_test,y_test,colors='red')
```
