```python
## scaling -- changew the size of the image enlarging or reducing
## scaled = cv.2 resize(img, (width, height))
cv2.resize(img, (200, 200))
scaled  = cv2.resize(img, (0,0), fx=0.5, fy=0.5)  # scale down by 50%
cv2.imshow('Scaled Image', scaled)
```

```python
## TRANSFORMATIONS --changes the loccation of the image
import numpy as np
rows , cols  = img.shape[:2]
tx ,ty = 80, 50
M = np.float32([[1, 0, tx], [0, 1, ty]])
#1 -- keep the  x coordinate the same
#0 -- keep the y coordinate the same
#tx -- translation in x direction
translated_image = cv2.warpAffine(img, M, (cols, rows))
plt.imshow(translated_image)
##warpaffine -- applies an affine transformation to the image
```

```python
## rotation -- rotates the image by a certain angle
center = (cols // 2, rows // 2)
angle = 45  # angle in degrees
scale = 0.5  # scale factor 
#scale -- zoom in or zoom out
## 1 = original image ,<1 = zoom out, >1 = zoom in
M = cv2.getRotationMatrix2D(center, angle, scale)
rotated_image = cv2.warpAffine(img, M, (cols, rows))
plt.imshow(rotated_image)
```

```python
## flipping -- flips the image horizontally or vertically
flip_image = cv2.flip(img, 0)  # 1 for horizontal flip
##flip code: 0 -- vertical flip, 1 -- horizontal flip, -1 -- both
plt.imshow(flip_image)
```

```python
##cropping -- removes the unwanted parts of the image
crop_img = img[10:200, 100:300]  # y1:y2, x1:x2
plt.imshow(crop_img)
```

```python
##cropping -- removes the unwanted parts of the image
crop_img = img[10:200,:]  # y1:y2, x1:x2
plt.imshow(crop_img)
```

```python
## rgb to grayscale conversion
img = cv2.imread('image1.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#gray = o.3r + 0.59g + 0.11b
plt.subplot(1,2,1)
plt.imshow(img)
plt.title('Original Image')
plt.subplot(1,2,2)
plt.imshow(gray , cmap='gray')
plt.title('Grayscale Image')
plt.show()
```

```python
#HISTOGRAM EQUALIZATION
## distribu;tion of pixel values in an image
color = ('blue', 'green', 'red')
for i , col in enumerate(color):
    #i = channel index (0 for blue, 1 for green, 2 for red)
    #none = mask (if any)-- full image histogram bna hai
    #256 = number of bins in the histogram
    # [0, 256] = range of pixel values
    hist = cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(hist, color=col)
    plt.xlim([0, 256])
    plt.show()
```
